#6.  What are the differences between NumPy arrays and Python lists ?
- Performance – NumPy arrays are faster due to optimized C-based implementation, while Python lists are slower due to dynamic typing.
- Memory Usage – NumPy arrays use less memory as they store elements in a contiguous block, whereas Python lists store references, using more memory.
- Data Type – NumPy arrays are homogeneous (all elements must be the same type), while Python lists can store mixed data types.
- Operations – NumPy supports vectorized operations for efficient computations, while Python lists require loops for element-wise operations.
- Functionality – NumPy provides built-in mathematical functions (np.mean(), np.sum()), whereas Python lists require manual implementation.
- Multi-dimensional Support – NumPy supports multi-dimensional arrays (ndarray), while Python lists need nested structures for multi-dimensional data.

#8. What does the term “vectorized operation” mean in NumPy ?

#13 A Why is handling missing data important in Pandas ?

- Here’s why it matters:
    - Ensures Data Integrity – Missing values can distort statistical calculations and summaries (e.g., mean, median, standard deviation).

#17 Explain the difference between apply() and map() in Pandas.
- Scope of Use
    - apply() works on both Series and DataFrames.
    - map() works only on Series.
- Function Application:
    - apply() can apply functions to entire rows or columns in a DataFrame.
    - map() applies functions element-wise on a Series.
- Complexity of Operations:
    - apply() supports more complex operations, including lambda functions and custom functions.
    - map() is primarily used for simpler element-wise transformations.
- Output Type:
    - apply() returns either a Series or a DataFrame depending on the operation.
    - map() always returns a Series.

#21 Why is NumPy’s array slicing faster than Python’s list slicing ?
- Memory Efficiency:NumPy slices return views, not copies, meaning they reference the original data instead of creating new objects.
- Contiguous Memory Storage:NumPy arrays use contiguous memory blocks, enabling faster access and computation, unlike Python lists, which store elements as separate objects with pointers.
- Optimized C Implementation:NumPy is implemented in C, allowing low-level optimizations that make slicing operations significantly faster than Python’s built-in lists.
- Vectorized Operations:Slicing in NumPy benefits from vectorized computations, leveraging SIMD (Single Instruction, Multiple Data) for performance boosts.

```python
#4.  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap ?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {'A': [1, 2, 3, 4, 5],
        'B': [2, 4, 6, 8, 10],
        'C': [5, 3, 2, 4, 1]}

df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Show the plot
plt.title("Correlation Matrix Heatmap")
plt.show()
```
